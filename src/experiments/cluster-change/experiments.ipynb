```python
def show_result(array, cluster_result):
    # show result
    fig = px.scatter(
        x=array[:, 0],
        y=array[:, 1],
        color=cluster_result,

    )
    fig.show()
```

# データの読み込み


```python
# MNIST
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import StandardScaler

# MNIST をロード
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
```

```python

# サンプリング
sample_num = 5000

X = mnist.data[:sample_num]      # shape: (70000, 784)
y = mnist.target[:sample_num]    # shape: (70000,)

# 高次元データ (784次元)
high_dim_data = StandardScaler().fit_transform(X)  # スケーリングしておくとPCAなどに便利

print("Shape of high_dim_data:", high_dim_data.shape)  # -> (70000, 784)
```

## 

```python
# pca
pca = PCA(n_components=2)
pca_result = pca.fit_transform(high_dim_data)

show_result(pca_result, y)

# t-SNE
tsne = TSNE(n_components=2, random_state=0, perplexity=30)
tsne_result = tsne.fit_transform(high_dim_data)
show_result(tsne_result, y)





```

```python
# 特定のクラス
n = 4
# データのフィルタ

mask = y == str(n)
filtered_data = high_dim_data[mask]
filtered_labels = y[mask]

# PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(filtered_data)
show_result(pca_result, filtered_labels)


# t-SNE
# tsne = TSNE(n_components=2, random_state=0, perplexity=30)
# tsne_result = tsne.fit_transform(filtered_data)
# show_result(tsne_result, filtered_labels)

# cpca
from contrastive import CPCA

cpca = CPCA()
cpca_result = cpca.fit_transform(filtered_data, X)
show_result(cpca_result[0], filtered_labels)


```
